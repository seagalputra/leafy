```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2gray
from skimage.transform import rescale
from skimage.feature import hog
from skimage.feature import daisy
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import cv2
import os
import glob
%matplotlib inline
```

## Load Dataset

```python
def load_dataset(path):
    folders = glob.glob(path)
    imagenames_list = []
    label = []
    for folder in folders:
        for f in glob.glob(folder+'/*.JPG'):
            folder_name = os.path.basename(folder)
            label_name = folder_name.split("_")
            label.append(label_name[1])
            imagenames_list.append(f)
    
    image_data = []
    
    for image in imagenames_list:
        image_data.append(io.imread(image))
    
    return image_data, label

folder_path = '../data/*'
image_data, label = load_dataset(folder_path)
```

## Image Descriptor

### Histogram of Oriented Gradient (HoG)

```python
def preprocessing(image):
    gray_image = rgb2gray(image)
    gray_image = rescale(gray_image, 0.5)
    return gray_image
```

```python
def hog_descriptor(image):
    fd, hog_image = hog(image, orientations=8, pixels_per_cell=(16, 16), 
                        cells_per_block=(1, 1), visualize=True, multichannel=False)
    return fd, hog_image
```

```python
gray_data = []
for i in range(len(image_data)):
    print('Processing image - %s' %i)
    gray_data.append(preprocessing(image_data[i]))
```

```python
plt.imshow(gray_data[0], cmap=plt.cm.gray)
plt.title(label[0])
plt.show()
```

```python
fd = []
fd_img = []
for image in gray_data:
    temp_fd, temp_image = hog_descriptor(image)
    fd.append(temp_fd)
    fd_img.append(temp_image)

fd = np.asarray(fd)
fd_img = np.asarray(fd_img)
```

```python
plt.imshow(fd_img[0], cmap=plt.cm.gray)
plt.title(label[0])
plt.show()
```

### Daisy

```python
def daisy_descriptor(image):
    descs, descs_img = daisy(image, step=180, radius=58, rings=2, histograms=6, 
                             orientations=8, visualize=True)
    return descs, descs_img
```

```python
fd = []
fd_img = []
for image in gray_data:
    temp_descs, temp_descs_img = daisy_descriptor(image)
    descs.append(temp_descs)
    descs_img.append(temp_descs_img)
    
descs = np.asarray(descs)
descs_img = np.asarray(descs_img)
```

## Classifier

### Shuffle Dataset

```python
shuffle_index = np.random.permutation(len(image_data))
fd, label = fd[shuffle_index], label[shuffle_index]
```

### Classify Image

```python
data_train, data_test, label_train, label_test = fd[:343], fd[343:], label[:343], label[343:]
```

```python
clf_knn = KNeighborsClassifier(n_neighbors=3, weights='distance', n_jobs=-1)
```

```python
scores = cross_val_score(clf_knn, data_train, label_train, cv=3, n_jobs=-1)
print('Cross Validation score : {}'.format(np.mean(scores)))
print(scores)
```

```python
clf_knn.fit(data_train, label_train)
```

```python
predict = clf_knn.predict(data_test)
```

```python
score = accuracy_score(label_test, predict)
print('Accuracy : {:.4%}'.format(score))
```
